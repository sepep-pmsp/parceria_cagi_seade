## Quero saber a quantidade de casamentos de municipio A pra municipio B

```python
colunas = ['codmunicipio_cja', 'codmunicipio_cjb']
```

```python
df = df[colunas]
```

```python
casamentos = {}
origem = 'codmunicipio_cja'
destino = 'codmunicipio_cjb'
for i, row in df.iterrows():

    cidade_origem = row[origem]
    cidade_destino = row[destino]
    if cidade_origem not in casamentos:
        casamentos[cidade_origem] = {}
    
    dici_cidade_origem = casamentos[cidade_origem]
    if cidade_destino not in dici_cidade_origem:
        dici_cidade_origem[cidade_destino] = 0

    dici_cidade_origem[cidade_destino]+=1
```

```python
df_casamentos = pd.DataFrame(casamentos)
```

```python
df_casamentos
```

```python
df_casamentos.loc[3550308, 3550308] = 0
```

```python
df_casamentos = df_casamentos.fillna(0)
```

```python
cidades_unicas = set(df[origem].unique()) | set(df[destino].unique())
```

```python
len(cidades_unicas)
```

```python
df_casamentos.head()
```

```python
df_casamentos.shape
```

```python
colunas = {}
for cidade in cidades_unicas:
    if cidade not in df_casamentos.columns:
        colunas[cidade] = [0] * len(df_casamentos)

df_colunas = pd.DataFrame(colunas)
```

```python
df_colunas.index = df_casamentos.index
```

```python
df_casamentos = pd.concat([df_casamentos, df_colunas], axis=1)
```

```python
len(df_casamentos)
```

```python
colunas = df_casamentos.columns
for cidade in cidades_unicas:
    if cidade not in df_casamentos.index:   
        df_casamentos.loc[cidade, colunas] = [0] * len(colunas)
```

```python
import geopandas as gpd
```

```python
shp_df = gpd.read_file('shape_municipios')
```

```python
shp_df.dtypes
```

```python
shp_df['CD_MUN_INT'] = shp_df['CD_MUN'].astype(int)
```

```python
nao_existe = set(df_casamentos.columns) - set(shp_df['CD_MUN_INT'].unique())
```

```python
df_casamentos.drop(list(nao_existe), axis=1, inplace=True)
df_casamentos.drop(list(nao_existe), axis=0, inplace=True)
```

```python
set(df_casamentos.columns) - set(shp_df['CD_MUN_INT'].unique())
```

```python
shp_df.head()
```

```python
shp_df.drop(['AREA_KM2', 'CD_MUN'], axis=1, inplace=True)
```

```python
shp_df.crs
```

```python
shp_df.geometry = shp_df.centroid
```

```python
shp_df = shp_df.to_crs(epsg=4326)
```

```python
shp_df
```

```python
shp_df['lon'] = shp_df.geometry.x 
shp_df['lat'] = shp_df.geometry.y
```

```python
shp_df.drop('geometry', axis=1, inplace=True)
```

```python
shp_df
```

```python
dados_mapa = []
columns = df_casamentos.columns
for mun_origem, row in df_casamentos.iterrows():
    lat_origem, long_origem = shp_df[shp_df['CD_MUN_INT']==mun_origem][['lat', 'lon']].values[0]
    for mun_destino in columns:
        lat_dest, long_dest = shp_df[shp_df['CD_MUN_INT']==mun_destino][['lat', 'lon']].values[0]

        qtd_casamentos = row[mun_destino]

        if qtd_casamentos>0:
            dados_par = {
                'S000' : qtd_casamentos,
                'lng_h' : long_origem,
                'lat_h' : lat_origem,
                'lng_w' : long_dest,
                'lat_w' : lat_dest,
                'mun_origem' : mun_origem,
                'mun_destino' : mun_destino
            }

            dados_mapa.append(dados_par)
```

```python
df_final = pd.DataFrame(dados_mapa)
```

```python
df_final.head()
```

```python
shp_df[shp_df['NM_MUN'].str.contains('Paulo')]
```

```python
cd_sao_paulo = 3550308
```

```python
df_final = df_final[(df_final['mun_origem']==cd_sao_paulo)|(df_final['mun_destino']==cd_sao_paulo)]
```

```python
"""
ArcLayer
========

Map of commutes to work within a segment of downtown San Francisco using a deck.gl ArcLayer.

Green indicates a start point, and red indicates the destination.

The data is collected by the US Census Bureau and viewable in the 2017 LODES data set: https://lehd.ces.census.gov/data/
"""

import pydeck as pdk


GREEN_RGB = [0, 255, 0, 40]
RED_RGB = [240, 100, 0, 40]

# Specify a deck.gl ArcLayer
arc_layer = pdk.Layer(
    "ArcLayer",
    data=df_final,
    get_width="S000 * 2",
    get_source_position=["lng_h", "lat_h"],
    get_target_position=["lng_w", "lat_w"],
    get_tilt=15,
    get_source_color=RED_RGB,
    get_target_color=GREEN_RGB,
    pickable=True,
    auto_highlight=True,
)

view_state = pdk.ViewState(latitude=-23.5505, longitude=-46.6333, bearing=45, pitch=50, zoom=8,)


TOOLTIP_TEXT = {"html": "{S000} casamentos <br /> Cidade do conjuge 1: {mun_origem}, <br /> Cidade do conjuge 2: {mun_destino}"}
r = pdk.Deck(arc_layer, initial_view_state=view_state, tooltip=TOOLTIP_TEXT)
r.to_html("arc_layer.html")
```
